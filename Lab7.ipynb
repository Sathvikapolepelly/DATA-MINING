```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data = {'Employee': ['John ', 'Sarah', 'Mike','Emma','Tom'],
    'Department': ['IT','HR','IT','Finance','HR'],
    'Salary': ['60000','50000','75000','65000','48000'],
    'Experience (Years)': ['5','3','7','4','2'],
    'Performance Rating (Out of 5)': ['4.5','3.8','4.9','4.2','3.5']}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df['Salary'] = df['Salary'].astype(int)
average_salary = df.groupby('Department')['Salary'].mean()
average_salary
```

```python
df['Performance Rating (Out of 5)'] = df['Performance Rating (Out of 5)'].astype(float)
high_performers = df[df['Performance Rating (Out of 5)'] > 4.0]
high_performers
```

```python
df['Increment'] = np.where(df['Performance Rating (Out of 5)'] > 4.0,  df['Salary'] * 0.10, df['Salary'] * 0.05)
df['Salary After Increment'] = df['Salary'] + df['Increment']
df
```

```python
sorted_df = df.sort_values(by='Salary After Increment', ascending=False)
sorted_df
```

```python
data = {
    "City": ["New York", "Los Angeles", "Chicago", "Houston", "Miami"],
    "Total Cases": [100000, 85000, 65000, 70000, 55000],
    "Recovered": [75000, 70000, 50000, 60000, 45000],
    "Deaths": [5000, 3500, 2500, 2000, 1500]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df['Recovery Rate (%)'] = (df['Recovered'] / df['Total Cases']) * 100
```

```python
df
```

```python
df['Death Rate (%)'] = (df['Deaths'] / df['Total Cases']) * 100
```

```python
df
```

```python
highcities = df[df['Death Rate (%)'] >3]
highcities
```

```python
sorted_recovery = df.sort_values(by='Recovery Rate (%)', ascending=False)
sorted_recovery
```

```python
total_cases = df['Total Cases'].sum()
print(total_cases)
```

```python
total_deaths = df['Deaths'].sum()
print(total_deaths)
```
